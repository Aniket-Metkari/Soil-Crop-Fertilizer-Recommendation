## Preprocessing using One-Hot Encoder

```python
y = fertilizer_df['Fertilizer Name'].copy()
X = fertilizer_df.drop('Fertilizer Name',axis=1).copy()
```

```python
ct = ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[3,4])],remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
X
```

## Train-test split

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,shuffle=True,random_state=42)
```

## Feature Scaling

```python
sc = StandardScaler()
print(X_train)
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

## Machine Learning Models

### Random Forest

```python
RFClassifier = RandomForestClassifier(n_estimators=200,criterion='gini',random_state=42)
RFClassifier.fit(X_train,y_train)
```

```python
y_pred = RFClassifier.predict(X_test)
score = accuracy_score(y_test,y_pred)
print("Random Forest Classifier's Accuracy is: ",score*100,"%")
print("\nClassification Report: \n\n",classification_report(y_test,y_pred))
```

### eXtreme Gradient Boosting (XGBoost)

```python
le = LabelEncoder()
XGBoostClassifier = XGBClassifier()
XGBoostClassifier.fit(X_train,y_train)
```

```python
y_pred = XGBoostClassifier.predict(X_test)
score = accuracy_score(y_test,y_pred)
print("XGBoost Classifier's Accuracy is: ",score*100,"%")
print("\nClassification Report: \n\n",classification_report(y_test,y_pred))
```

### LightGBM

```python
from lightgbm import LGBMClassifier
LightGBMClassifier = LGBMClassifier()
LightGBMClassifier.fit(X_train,y_train)
```

```python
y_pred = LightGBMClassifier.predict(X_test)
score = accuracy_score(y_test,y_pred)
print("LightGBM Classifier's Accuracy is: ",score*100,"%")
print("\nClassification Report: \n\n",classification_report(y_test,y_pred))
```

## LightGBM Model by Hyperparameter Tuning

```python
from lightgbm import LGBMClassifier
from sklearn.model_selection import GridSearchCV

# Create a LightGBM classifier with regularization

lgbm = LGBMClassifier(num_classes=len(y.unique()),objective ='multiclass', lambda_l1=0.1, lambda_l2=0.1, min_gain_to_split=0.1)

# Set up the hyperparameter grid for tuning
params = {
    "learning_rate": [0.01, 0.1, 0.5],
    "n_estimators": [50, 100, 200],
    "max_depth": [3, 5, 7],
    "subsample": [0.5, 0.7, 1.0],
    "colsample_bytree": [0.5, 0.7, 1.0]
}

# Set up the grid search with cross-validation
grid_search = GridSearchCV(lgbm, params, scoring="accuracy", cv=5)

# Fit the grid search to the training data
grid_search.fit(X_train, y_train)

# Print the best hyperparameters
print("Best hyperparameters:", grid_search.best_params_)

# Use the best model to make predictions on the testing data
y_pred = grid_search.predict(X_test)

# Calculate the accuracy score
accuracy = accuracy_score(y_test, y_pred)
temp = cross_val_score(grid_search, X_test,y_test,cv=5)
class_report = classification_report(y_test,y_pred)

print("Accuracy: ",accuracy)
print("\nClassification Report: \n\n",class_report)
```

```python
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from lightgbm import LGBMClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
import numpy as np
# Assuming X_train, X_test, y_train, y_test are already defined

# Instantiate the classifiers
lgbm = LGBMClassifier(num_classes=len(np.unique(y_train)), objective='multiclass', lambda_l1=0.1, lambda_l2=0.1, min_gain_to_split=0.1)
rf = RandomForestClassifier(n_estimators=100, random_state=42)

# Fit the classifiers
lgbm.fit(X_train, y_train)
rf.fit(X_train, y_train)

# Make predictions
lgbm_pred = lgbm.predict(X_test)
rf_pred = rf.predict(X_test)

# Combine predictions using a VotingClassifier
voting_clf = VotingClassifier(estimators=[('lgbm', lgbm), ('rf', rf)], voting='hard')
voting_clf.fit(X_train, y_train)
voting_pred = voting_clf.predict(X_test)

# Calculate the accuracy scores
lgbm_score = accuracy_score(y_test, lgbm_pred)
rf_score = accuracy_score(y_test, rf_pred)
voting_score = accuracy_score(y_test, voting_pred)

# Print the accuracy scores
print("LightGBM Classifier's Accuracy: ", lgbm_score * 100, "%")
print("Random Forest Classifier's Accuracy: ", rf_score * 100, "%")
print("Voting Classifier's Accuracy: ", voting_score * 100, "%")

# Print the classification reports
print("\nClassification Report for LightGBM Classifier:\n", classification_report(y_test, lgbm_pred))
print("\nClassification Report for Random Forest Classifier:\n", classification_report(y_test, rf_pred))
print("\nClassification Report for Voting Classifier:\n", classification_report(y_test, voting_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
from lightgbm import LGBMClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
import numpy as np

# Assuming X_train, X_test, y_train, y_test are already defined

# Instantiate the classifiers
lgbm = LGBMClassifier(num_classes=len(np.unique(y_train)), objective='multiclass', lambda_l1=0.1, lambda_l2=0.1, min_gain_to_split=0.1)
rf = RandomForestClassifier(n_estimators=100, random_state=42)

# Fit the classifiers
lgbm.fit(X_train, y_train)
rf.fit(X_train, y_train)

# Predict probabilities for each class
lgbm_probs = lgbm.predict_proba(X_test)
rf_probs = rf.predict_proba(X_test)

# Combine the predicted probabilities using average
average_probs = (lgbm_probs + rf_probs) / 2

# Convert probabilities back to class labels using a threshold
threshold = 0.5  # Adjust this threshold if necessary
y_pred_avg = np.argmax(average_probs, axis=1)

# Calculate the accuracy score
score_avg = accuracy_score(y_test, y_pred_avg)
print("Averaged Classifier's Accuracy is: ", score_avg * 100, "%")

# Generate the classification report
print("\nClassification Report: \n\n", classification_report(y_test, y_pred_avg))
```
